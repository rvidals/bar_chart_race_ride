```python
# Importar bibliotecans necessárias
import os
import glob
import pandas as pd
# Bibliteca responsavel pelos gráficos
import bar_chart_race as bcr 
```

```python
#Caminho do arquivo xlsx
os.chdir('..') #Voltar uma pasta
path = os.path.join(os.getcwd(),r'DADOS')
file = glob.glob(path + '/Modelo.csv')
```

```python
#Salvar em um dataframe
df = pd.read_csv(file[0],
                 encoding='utf-8',
                 sep=';')
```

```python
#Organizar o DF: 
        #1) Municípios 
        #2) Série Histórica ordenada

df = df[['Municípios',
        '1960',
        '1970',
        '1980',
        '1991',
        '2000',
        '2010',
        '2014',
        '2021',
        '2022'
       ]]
```

```python
#Criar uma cópia para trabalha-lá
df_copy = df.copy()
```

```python
"""
Transpor o DataFrame:
1) O Ano ficará como o index do dataframe transposto
2) As colunas ficaram sem nome. Esse problema será tratado mais a frente
"""
df_t = df_copy.T
```

```python
df_t
```

```python
#Renomear as colunas com os valores da primeira linha. 
#Ao renomear será removido espaços nos nomes dos municípios
df_t = df_t.rename(columns=df_t.iloc[0].str.strip())
#Resolver problemas de formatação do mun São João d'Aliança
df_t = df_t.rename(columns={'São João dAliança':"São João d'Aliança"})
```

```python
#Remover a primeira linha do dataframe
df_t = df_t.iloc[1:,:]
```

```python
#Colocar um nome no index
df_t.index.name='Ano'
```

```python
df_t
```

```python
#Remover caracteres especiais
df_t = df_t.replace(
    {
        "-":"",
        " ":""
    },
    regex=True

)
df_t
```

```python
df_t.info()
```

```python
#Transformar os valores para numéricos
df_t = df_t.apply(pd.to_numeric)
```

```python
df_t.info()
```

```python
#Com o Distrito Federal
df_tComDF = df_t 
#Sem o Distrito Federal
df_tSemDF = df_t[['Cidade Ocidental', 'Valparaíso de Goiás', 'Cocalzinho de Goiás', 'Alto Paraíso de Goiás',
 'Planaltina', 'Águas Lindas de Goiás', 'Cristalina', 'Padre Bernardo', 'Barro Alto', 'Alexânia', 
 'Goianésia', 'Cavalcante',  'Pirenópolis', 'Luziânia', 'Formosa', 'Unaí', 'Buritis', "São João d’Aliança",
 'Santo Antônio do Descoberto', 'Vila Propício', 'Água Fria de Goiás', 'Arinos', 'Alvorada do Norte', 'Novo Gama',
 'Cabeceira Grande', 'Cabeceiras', 'Abadiânia', 'Flores de Goiás', 'Mimoso de Goiás', 'Corumbá de Goiás',
 'Simolândia', 'Niquelândia', 'Vila Boa']]
```

```python
#Funções com parâmetros para melhorar visualização do gráfico
def summary(values, ranks):
    total_habitantes = values.sum()
    s = f"Total de habitantes: {total_habitantes:_.0f}"
    s = s.replace('.',',').replace('_','.')
    return {'x': .75, 'y': .05, 's': s, 'size': 8}

def func(values, ranks):
    return values.quantile(.9)
```

```python
#Com Distrito Federal
bcr.bar_chart_race(df_tComDF,
                   os.path.join(os.getcwd(),'GIF') + '\Crescimento_Populacional_Ride_DF.mp4',
                   steps_per_period=10,
                   perpendicular_bar_func=func, 
                   figsize=(9,7),
                   dpi=288,
                   period_summary_func = summary,
                   period_length=1750, 
                   fixed_max=True)
```

```python
#Sem Distrito Federal
bcr.bar_chart_race(df_tSemDF,
                   os.path.join(os.getcwd(),'GIF') + '\Crescimento_Populacional_Ride.mp4',
                   steps_per_period=10,
                   perpendicular_bar_func=func, 
                   figsize=(9,7),
                   dpi=288,
                   period_summary_func = summary,
                   period_length=1750, 
                   fixed_max=True)
```
